# **Que :- 1. What is the difference between a function and a method in Python?**

**Ans:-**

In Python, both functions and methods are blocks of code that perform a specific task, but they differ primarily in their context and usage.

# **1.Function:-**
**Definition:** A function is a reusable piece of code that performs a specific task and can be defined using the **'def'** keyword.




**Usage:** Functions can be called independently and are not associated with any object. They operate on data that is passed to them.

**Example:**

```python
def add(a, b):
    return a + b

result = add(5, 3)  # result is 8
print(result)  # This will output the result
```

# **2.Method:-**

**Definition:** A method is a function that is associated with an object and is defined within a class. It is called on instances of that class and can access and modify the objectâ€™s attributes.

**Usage:** Methods are called on objects, and they usually operate on the data contained within the object (the instance attributes).

**Example:**


```python
class Calculator:
    def __init__(self, value):
        self.value = value

    def add(self, number):
        self.value += number

calc = Calculator(10)
calc.add(5)  # calc.value is now 15
print(calc.value)  # This will output the updated value
```

**Key Differences**

**1.Context:**

**Function**: Defined independently of any class.

**Method:** Defined within a class and operates on the instance of the class.

**2.Invocation:**

**Function:** Called by its name directly, e.g., add(5, 3).

**Method:** Called on an object of the class, e.g., calc.add(5).

**3.Access to Data:**

**Function:** Operates only on the parameters passed to it.

**Method:** Can access and modify the attributes of the class instance (self).

In summary, methods are a special type of function that are bound to the objects of a class and have access to the object's state. Functions are more general and are not tied to any object.

# **Que:-2 Explain the concept of function arguments and parameters in Python.**

Ans:-
In Python, function arguments and parameters are fundamental concepts that define how functions receive and use data. Here's a detailed explanation of each:

**1.Function Parameters:**

**Parameters** are variables listed in the function definition that serve as placeholders for the values that will be passed to the function. They define what kind of input the function expects. Parameters are used to accept input values when the function is called.

**Defination:-** Parameters are defined in the parentheses following the function name in the function declaration.

Example:

```python
greet("Alice", 30)
def greet(name, age):
    print(f"Hello, {name}! You are {age} years old.")
```

In this example, name and age are parameters. They specify that the greet function expects two inputs.

**2.Function Arguments:-**
Arguments are the actual values or expressions passed to the function when it is called. These values are assigned to the corresponding parameters in the function definition.

**Definition:** Arguments are provided in the function call and are matched with the parameters in the function definition.

Example:



```python
greet("Alice", 30)
```

In this call, "Alice" and 30 are arguments. They are passed to the greet function, where "Alice" is assigned to the name parameter and 30 is assigned to the age parameter.

**How They Work Together**

When you call a function, the values you provide as arguments are used to replace the parameters in the function definition. This allows the function to operate with the specific values provided.

**Types of Parameters and Arguments**

**1.Positional Parameters and Arguments:-**

**Positional Parameters:** These parameters must be provided in the exact order they are defined in the function.

**Positional Arguments:** The values passed must match the order of the parameters.
Example:

```python
def divide(x, y):
    return x / y

result = divide(10, 2)  # x=10, y=2
```

**2.Keyword Parameters and Arguments:-**

**Keyword Parameters**: Parameters that are specified by name in the function call. This allows you to provide arguments in any order.

**Keyword Arguments**: Arguments that specify the parameter name.

Example:

```python
def describe(name, age):
    print(f"{name} is {age} years old.")

describe(age=25, name="John")
```

**3.Default Parameters:-**

Parameters can have default values, which are used if no argument is provided for those parameters.

Example:

```python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Alice")          # Uses default greeting "Hello"
greet("Bob", "Hi")      # Uses specified greeting "Hi"
```

**4.Variable-Length Parameters:-**

# **args:**
 Allows the function to accept any number of positional arguments, which are stored in a tuple.

```python
def sum_all(*numbers):
    return sum(numbers)

print(sum_all(1, 2, 3, 4))  # Output: 10
```

 Allows the function to accept any number of keyword arguments, which are stored in a dictionary.

Example:

```python
def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30)  # Outputs: name: Alice \n age: 30
```

**Summary**

**Parameters** are variables listed in the function definition that represent what kind of inputs the function expects.

**Arguments** are the actual values or expressions you provide to the function when you call it.

**Different types of parameters** (e.g., positional, keyword, default, variable-length) allow functions to handle a variety of input scenarios, making your code more flexible and reusable.

# **Que:-4 What is the purpose of the `return` statement in a Python function?**

**Ans:-** The **return statement** in a Python function serves several important purposes:

**1. Outputting a Result**
The primary purpose of the return statement is to send a result from the function back to the caller. When a function is called, it can perform computations or operations and then use return to provide a value as the output of the function.

Example:

```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)  # Output: 8
```

In this example, return a + b sends the result of a + b back to the caller, which is then stored in the result variable and printed.

**2. Ending Function Execution**

The return statement also terminates the execution of the function. Once return is executed, the function stops running, and control is returned to the point where the function was called. Any code after the return statement in the function will not be executed.

Example:

```python
def greet(name):
    if not name:
        return "No name provided"
    return f"Hello, {name}!"

print(greet(""))  # Output: No name provided
print(greet("Alice"))  # Output: Hello, Alice!
```

Here, if name is an empty string, return "No name provided" ends the function early, skipping the remaining code.

**3. Returning Multiple Values**

Python functions can return multiple values as a tuple. This is useful for functions that need to return more than one result.

Example:

```python
def min_max(numbers):
    return min(numbers), max(numbers)

low, high = min_max([1, 2, 3, 4])
print(low, high)  # Output: 1 4
```

In this example, return min(numbers), max(numbers) returns a tuple containing the minimum and maximum values, which can then be unpacked into separate variables.

**4. Returning None**

If a function does not explicitly use the return statement, or if the return statement does not include an expression, the function returns None by default. This is often used to indicate that the function does not produce a meaningful result or is used for its side effects (e.g., printing to the console).

Example:

```python
def print_message(message):
    print(message)

result = print_message("Hello!")
print(result)  # Output: Hello! \n None
```

In this example, print_message prints the message but does not return any value, so result will be None.

**Summary**
The return statement in Python functions:

**1.Provides Output:** It allows a function to send a result back to the caller.

**2.Ends Execution:** It terminates the function's execution and returns control to the caller.

**3.Supports Multiple Returns:** It can return multiple values as a tuple.

**4.Default Return:** Functions return None if no return statement is provided or if return is used without an expression.

Understanding the return statement is crucial for creating functions that are both functional and effective, allowing you to produce and manage results efficiently.

# **Que:-5 What are iterators in Python and how do they differ from iterables?**

**Ans:-**In Python, iterators and iterables are closely related concepts that play a crucial role in managing and working with sequences of data. Hereâ€™s a detailed explanation of both, including how they differ:

**Iterables**
**Definition:** An iterable is any Python object that can return its members one at a time, allowing it to be iterated over in a loop. Iterables implement the __iter__() method, which returns an iterator.

**Common Examples:** Lists, tuples, strings, dictionaries, and sets are examples of iterable objects.

**How It Works:** An iterable can be used in a for loop or any context that requires iteration. When you use an iterable in a loop, Python implicitly calls the iter() function on it, which returns an iterator.

Example:



```python
numbers = [1, 2, 3, 4]

# This list is an iterable
for number in numbers:
    print(number)
```

**Iterators**
**Definition**: An iterator is an object that represents a stream of data. It implements two methods:

__iter__(): This method returns the iterator object itself. It is required to implement the iterator protocol.

__next__(): This method returns the next value from the stream. When there are no more items to return, it raises the StopIteration exception to signal that the iteration is complete.

**How It Works:** Iterators are used to iterate through elements one at a time. They maintain their state between iterations, so they keep track of where they are in the data stream.

Example:

```python
# Create an iterator from an iterable
numbers = [1, 2, 3, 4]
iterator = iter(numbers)

# Use the iterator to iterate over elements
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
```

```python
#If you try to get the next value after the last item, youâ€™ll get a StopIteration exception:
try:
    print(next(iterator))  # Output: 3
    print(next(iterator))  # Output: 4
    print(next(iterator))  # Raises StopIteration
except StopIteration:
    print("No more items")
```

**Key Differences**

**1.Definition and Usage:**

**Iterable**: An object that can be iterated over, such as a list or string. It defines how to return an iterator via the __iter__() method.

**Iterator:** An object that performs the actual iteration. It defines how to get the next item via the __next__() method and maintains the state of iteration.

**2.Methods:**

**Iterable:** Implements __iter__(), which returns an iterator.

**Iterator**: Implements both __iter__() (returns itself) and __next__() (returns the next item).

**3.State Management:**

**Iterable:** Does not keep track of the iteration state.
**Iterator:** Maintains the state of where it is in the iteration process.

**4.Example Usage:**

**Iterable:**

```python
numbers = [1, 2, 3]
for number in numbers:
    print(number)
```

**Iterator:**

```python
iterator = iter(numbers)
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
```

# **Custom Iterators:-**

You can create your own iterators by defining a class with __iter__() and __next__() methods.

Example:

```python
class Reverse:
    def __init__(self, data):
        self.data = data
        self.index = len(data)

    def __iter__(self):
        return self

    def __next__(self):
        if self.index == 0:
            raise StopIteration
        self.index = self.index - 1
        return self.data[self.index]

rev = Reverse('giraffe')
for char in rev:
    print(char)
```

In this example, the Reverse class creates an iterator that iterates over a sequence in reverse order.

Understanding iterables and iterators allows you to leverage Python's powerful iteration constructs and create efficient, clean code that can handle large datasets and complex sequences.






# **Que:-6 Explain the concept of generators in Python and how they are defined.**

Ans:- Generators in Python are a special type of iterator that simplify the creation of iterators by allowing you to write functions that produce a sequence of results lazily, one at a time, without storing the entire sequence in memory. Generators are defined using functions with the yield keyword instead of return. Hereâ€™s a detailed explanation of generators, including how they are defined and used:

**What is a Generator?**

A generator is a type of iterable that generates values on-the-fly as you iterate over them. Unlike lists or other sequences that store all their values in memory, generators compute their values dynamically, which can be more memory efficient, especially for large datasets or streams of data.

**How Generators Work**

**1.Lazy Evaluation:** Generators produce values one at a time and only when requested. They do not compute or store all values at once.

**2.Stateful Iteration:** Generators maintain their state between iterations, allowing them to pick up where they left off each time next() is called.

**3.Yielding Values:** Instead of returning all values at once, a generator uses the yield statement to yield one value at a time.

**4.StopIteration:** When the generator is exhausted (i.e., no more values to yield), it raises the StopIteration exception to signal the end of the iteration.

**Defining a Generator**

To define a generator, use the yield keyword in a function. The function with yield is called a generator function. Each call to the generator function returns an iterator object.

Example: Basic Generator Function

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1
```

In this example, count_up_to is a generator function that yields numbers from 1 to max.

**Using a Generator**

You can use a generator just like any other iterator, typically with a for loop or by manually calling next().

**Example 1: Iterating with a for Loop**

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1
for number in count_up_to(5):
    print(number)

```

In this example, the for loop iterates over the sequence of numbers from 1 to 5, yielding one number at a time.

**Example 2:**  **Using next() to Manually Retrieve Values**

You can manually retrieve values from the generator using the next() function. This approach gives you control over when to retrieve each value and handles the StopIteration exception when the generator is exhausted.

```python
gen = count_up_to(3)

print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3

try:
    print(next(gen))  # Raises StopIteration
except StopIteration:
    print("No more items")
```

In this example, the generator gen produces values 1, 2, and 3, and then raises a StopIteration exception when there are no more items.

**Example 3**: **Converting to a List**

You can convert the generator to a list to retrieve all its values at once. This is useful for inspecting or using all the values produced by the generator.

```python
numbers = list(count_up_to(4))
print(numbers)
```

In this example, the list() function consumes the generator and collects all the yielded values into a list.

**Example 4: Using the Generator in a Function**

You can also use the generator within another function or context where you need a sequence of values.

```python
def process_numbers():
    for number in count_up_to(3):
        print(f"Processing number: {number}")

process_numbers()
```

In this example, the process_numbers function calls the generator and processes each value it produces.

These examples demonstrate how to effectively use the count_up_to generator function to produce and work with sequences of numbers. Generators are powerful tools in Python for handling sequences efficiently and elegantly.

# **Que:- 7 What are the advantages of using generators over regular functions?**

Generators offer several advantages over regular functions, especially when dealing with large datasets or streams of data. Hereâ€™s a detailed look at the benefits:

**1. Memory Efficiency**

**Advantage:** Generators produce values one at a time and do not store the entire sequence in memory.

**Benefit:** This is particularly useful when working with large data sets or sequences where loading all data into memory would be impractical or impossible.

**Example:** Reading a large file line by line with a generator is much more memory-efficient than reading the entire file at once.

```python
with open('large_file.txt', 'w') as f:
    f.write("This is line 1\n")
    f.write("This is line 2\n")
    f.write("This is line 3\n")
    f.write("This is line 4\n")
    f.write("This is line 5\n")

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

print("Reading lines from the file:")
for line in read_large_file('large_file.txt'):
    print(line)
```

**2. Lazy Evaluation**

**Advantage:** Generators use lazy evaluation, meaning they compute values only when needed.

**Benefit**: This allows for on-demand computation of values, which can improve performance and reduce the time required to produce results.

**Example**: Generating the Fibonacci sequence up to a certain number only computes values as they are requested, rather than all at once.

```python
def fibonacci(limit):
    a, b = 0, 1
    while a <= limit:
        yield a
        a, b = b, a + b

print("Using a for loop:")
for number in fibonacci(10):
    print(number)

print("\nUsing next():")
fib_gen = fibonacci(10)
try:
    while True:
        print(next(fib_gen))
except StopIteration:
    pass

print("\nConverting to a list:")
fib_numbers = list(fibonacci(10))
print(fib_numbers)
```

**3. Reduced Overhead**

**Advantage:** Generators avoid the overhead of creating and managing large collections of data.

**Benefit:** This can result in reduced memory usage and faster execution for operations that involve a large number of elements.

**Example:** A generator for even numbers can be used without having to store all the numbers in a list.

```python
def even_numbers(limit):
    for num in range(2, limit + 1, 2):
        yield num

print("Using a for loop:")
for number in even_numbers(10):
    print(number)

print("\nUsing next():")
even_gen = even_numbers(10)
try:
    while True:
        print(next(even_gen))
except StopIteration:
    pass

print("\nConverting to a list:")
even_numbers_list = list(even_numbers(10))
print(even_numbers_list)
```

**4. Simplified Code**

**Advantage**: Generators can simplify code by eliminating the need for manual state management and iteration control.

**Benefit**: This leads to cleaner and more readable code that focuses on generating values rather than managing iteration logic.

**Example**: Generators can replace complex iterator classes with simpler generator functions.

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

print("Using a for loop:")
for number in countdown(5):
    print(number)

print("\nUsing next():")
countdown_gen = countdown(5)
try:
    while True:
        print(next(countdown_gen))
except StopIteration:
    pass

print("\nConverting to a list:")
countdown_list = list(countdown(5))
print(countdown_list)
```

**5. Infinite Sequences**

**Advantage**: Generators can produce infinite sequences, which would be impractical to represent with finite collections.

**Benefit**: This allows for the generation of an unbounded series of values without needing to handle or store them all at once.

**Example**: A generator that produces an infinite sequence of natural numbers.



```python
def count_up(start=1):
    while True:
        yield start
        start += 1

print("Using a for loop:")
# Using a for loop to print the first 5 numbers
for _ in range(5):
    print(next(count_up()))

print("\nUsing next():")
count_up_gen = count_up()
try:
    for _ in range(5):
        print(next(count_up_gen))
except StopIteration:
    pass
```

**6. Efficient Pipeline Processing**

**Advantage**: Generators are well-suited for building data processing pipelines where data flows through multiple stages.

**Benefit**: This allows for efficient, step-by-step transformation of data without intermediate storage.

**Example**: A pipeline for processing data in stages, such as filtering and mapping.

```python
def filter_even(numbers):
    for num in numbers:
        if num % 2 == 0:
            yield num

def square_numbers(numbers):
    for num in numbers:
        yield num * num

pipeline = square_numbers(filter_even(range(1, 10)))
print(list(pipeline))  # Output: [4, 16, 36, 64]
```

**Summary**

**Advantages of Generators:**

**1.Memory Efficiency**: They generate values one at a time, avoiding the need to store the entire sequence in memory.

**2.Lazy Evaluation**: They compute values only as needed, which can lead to better performance.

**3.Reduced Overhead**: They eliminate the need for managing large collections, leading to reduced memory usage and overhead.

**4.Simplified Code:** They simplify code by handling iteration internally and eliminating the need for manual state management.

**5.Infinite Sequences:** They can handle and generate infinite sequences without requiring infinite memory.

**6.Efficient Pipelines:** They facilitate efficient data processing pipelines, allowing data to be processed in stages without intermediate storage.

Generators provide a powerful and flexible way to handle sequences and data processing, making them a valuable tool in Python programming.


# **Que:-8 What is a lambda function in Python and when is it typically used?**

**Ans:-** A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions defined with def, lambda functions are concise and are usually written in a single line.

**Syntax**

The syntax of a lambda function is:

```python
lambda arguments: expression
```

**1.lambda:** The keyword used to define the lambda function.

**2.arguments:** A list of parameters (similar to those in a regular function).

**3.expression:** A single expression that is evaluated and returned. It cannot contain multiple expressions or statements.

**Example:-**
Here is a basic example of a lambda function:

```python
# Regular function definition
def add(x, y):
    return x + y

# Equivalent lambda function
add_lambda = lambda x, y: x + y

print(add(2, 3))        # Output: 5
print(add_lambda(2, 3)) # Output: 5
```

**Typical Uses**
Lambda functions are typically used in the following scenarios:

**1.As Arguments to Higher-Order Functions:**
Lambda functions are often used as arguments to functions like map(), filter(), and sorted().

*   **map():** Applies a function to all items in an iterable and returns a map object.







```python
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x ** 2, numbers))
print(squares) # Output: [1, 4, 9, 16, 25]
```



*  **filter():** Filters items in an iterable based on a function and returns a filter object.




```python
numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers) # Output: [2, 4]
```



*   **sorted():** Sorts an iterable and can use a lambda function to determine the sorting criteria.



```python
data = [(1, 'apple'), (3, 'banana'), (2, 'cherry')]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data) # Output: [(1, 'apple'), (3, 'banana'), (2, 'cherry')]
```

**2.In Short-lived Operations:**
Lambda functions are useful when you need a simple function for a short period and don't want to define a full function using def.

```python
add = lambda x, y: x + y
print(add(10, 5))  # Output: 15
```

**3.In Functional Programming Constructs:**
Lambda functions are used in functional programming constructs like function composition, currying, and more.

```python
def make_incrementer(n):
    return lambda x: x + n

increment_by_5 = make_incrementer(5)
print(increment_by_5(10)) # Output: 15
```

**Limitations**

* Single Expression: Lambda functions are limited to a single expression. They cannot contain multiple expressions or statements, which makes them less suitable for complex operations.

* Readability: Overuse of lambda functions, especially for complex operations, can reduce code readability.

In summary, lambda functions in Python are a concise way to define small, one-off functions and are commonly used for short-lived operations or as arguments to higher-order functions.






# **Que:-9 Explain the purpose and usage of the `map()` function in Python**

**Ans:-**
The **map()** function in Python is a built-in function that allows you to apply a specified function to each item in an iterable (such as a list, tuple, or set) and return an iterable (specifically, a map object) containing the results. Itâ€™s a powerful tool for transforming data in a functional programming style.

**Purpose:-**
The primary purpose of map() is to apply a given function to each item in an iterable and produce a new iterable with the results. This can be useful for data transformation, processing, and applying operations across a sequence of elements without explicitly writing a loop.

**Syntax:-** **map(function, iterable, ...)**

* '**function**': A function that will be applied to each item in the iterable. This function can be a regular function, lambda function, or any callable object.

* **iterable**: An iterable (e.g., list, tuple) whose elements are to be processed by the function.

* You can also pass multiple iterables to map(). In this case, the function must accept as many arguments as there are iterables.

**Example Usage**

**1.Basic Example with a Single Iterable**

Hereâ€™s a simple example that demonstrates how to use map() to square each number in a list:


```python
def square(x):
    return x ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)

print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]
```

In this example:

* **square** is a function that takes a number and returns its square.

* **map(square, numbers)** applies the **square** function to each element in the numbers list.

* **list(squared_numbers)** converts the **map** object to a list to print the results.

**2.Using a Lambda Function**

You can use a lambda function instead of defining a separate function:

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x ** 2, numbers)

print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]
```

**3.Multiple Iterables**

If you provide multiple iterables, the function must accept as many arguments as there are iterables. Hereâ€™s an example that adds corresponding elements from two lists:

```python
def add(x, y):
    return x + y

list1 = [1, 2, 3]
list2 = [4, 5, 6]
summed_lists = map(add, list1, list2)

print(list(summed_lists))  # Output: [5, 7, 9]
```

In this example:

* **add** is a function that takes two arguments and returns their sum.

* **map(add, list1, list2)** applies the add function to corresponding elements of list1 and list2.

**Characteristics**

* **Lazy Evaluation**: map() returns an iterator (a map object) that is lazily evaluated. This means the function is applied only when you iterate over the map object. To get results, you often need to convert the map object to a list or another iterable.

* **Efficiency**: Since map() processes items lazily, it can be more memory-efficient than using a list comprehension or a loop for large datasets.

**Summary**

The map() function is a convenient way to apply a function to each item in an iterable and can handle multiple iterables simultaneously. It promotes a functional programming style by avoiding explicit loops and can be combined with lambda functions for concise code.

# **Que:-10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**


**Ans:-** In Python, map(), reduce(), and filter() are built-in functions used for functional programming. They each serve distinct purposes for processing iterables. Hereâ€™s a breakdown of their differences and usage:

**1.map()**

**Purpose**:

Applies a function to each item in an iterable (or multiple iterables) and returns an iterable of the results.

**Syntax: map(function, iterable, ...)**

**Usage**:

Useful for transforming or processing data in an iterable. The function is applied to each element, and the results are collected into a new iterable.
Example:


```python
numbers = [1, 2, 3, 4, 5]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16, 25]
```

**2.reduce()**

**Purpose**:

* Applies a binary function (a function that takes two arguments) cumulatively to the items of an iterable, reducing the iterable to a single value.

**Syntax: reduce(function, iterable, [initializer])**

**Usage**:

Useful for performing aggregations or reductions, such as summing up values or concatenating items. It processes the iterable in a cumulative way, combining elements one by one.
Example:

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]
sum_result = reduce(lambda x, y: x + y, numbers)
print(sum_result)  # Output: 15
```

**3.filter()**

**Purpose**:

Filters items in an iterable based on a function that returns True or False. It returns an iterable of items for which the function returns True.

**Syntax: filter(function, iterable)**

**Usage**:

Useful for selecting elements from an iterable that meet a certain condition. The function should return True for elements you want to include and False for elements you want to exclude.

Example:

```python
numbers = [1, 2, 3, 4, 5]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4]
```

**Key Differences:-**

**1.Function Type:**

* **map()**: Applies a function to each item, resulting in a new iterable with transformed items.

* **reduce()**: Applies a function cumulatively to combine items into a single value.

* **filter()**: Applies a function to each item to filter the iterable, resulting in a new iterable with only the items that satisfy the condition.

**2.Return Value:**

* **map()**: Returns an iterable of results after applying the function to each item.
* **reduce()**: Returns a single cumulative result after combining all items.

* **filter()**: Returns an iterable of items that pass the test implemented by the function.

**3.Use Case:**

* **map()**: Used for transformations and modifications of data.
* **reduce()**: Used for aggregations or reductions of data to a single value.
* **filter()*: Used for selecting subsets of data based on a condition.

**Summary**

* **map()** is for applying a function to each item and getting a transformed iterable.
* **reduce()** is for reducing a sequence of items to a single value using a cumulative function.
* **filter()** is for filtering items from an iterable based on a condition.

Understanding these functions allows you to write more functional and declarative code, making it easier to process and transform data in Python.

## **Que:- 4 Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Example
number = 31  # Change this number to test other values
if 1 <= number <= 200:
    print(f"The number {number} is prime: {is_prime(number)}")
else:
    print("Number is out of the range 1 to 200.")
```

## **Que:- 5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example
num_terms = 10  # Change this to the desired number of terms
fibonacci_sequence = FibonacciIterator(num_terms)

print(f"Fibonacci sequence with {num_terms} terms:")
for number in fibonacci_sequence:
    print(number)
```

## **Que:- 7 Implement a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file_lines(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Yield each line, removing the trailing newline character
    except FileNotFoundError:
        print(f"The file at {file_path} was not found.")
    except IOError as e:
        print(f"An I/O error occurred: {e}")

# Example :
file_path = 'mohit.txt'  # Replace with the path to your file
for line in read_file_lines(file_path):
    print(line)
```

## **Que:- 10 Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
def is_not_vowel(char):
    vowels = 'aeiouAEIOU'
    return char not in vowels

def remove_vowels(input_string):
    # Use filter() to remove vowels
    filtered_chars = filter(is_not_vowel, input_string)
    # Join the filtered characters back into a string
    return ''.join(filtered_chars)

# Example:
input_string = "My name is Mohit."
result = remove_vowels(input_string)
print(f"String without vowels: {result}")
```
